```python
# Check if the directory exists
directory = r"C:\Users\santo\Repos\SSL-TIE_TFG\datasets\soundnet\Dataset\Data"
if not os.path.exists(directory):
    print(f"Directory {directory} does not exist.")
else:
    data = []
    count = 0
    for root, dirs, files in os.walk(directory):
        # print("Root: ",root,", dirs:",dirs,", files:",files)
        data.append(files)
        count += len(files)
    print(count)
    print(len(data),data)
```

```python
# Concatenate all the files
data_compressed = [item for sublist in data for item in sublist]


print(len(data_compressed),data_compressed)
```

```python
import pandas as pd
df = pd.DataFrame(data_compressed)

# Split the file extensions into another column
df['Filename'] = df[0].apply(lambda x: x.split('.')[0])
df['Extension'] = df[0].apply(lambda x: os.path.splitext(x)[1])
df['weird extension'] = df[0].apply(lambda x: os.path.splitext(os.path.splitext(x)[0])[1])

#All the '' change them to None
df['weird extension'] = df['weird extension'].apply(lambda x: None if x == '' else x)
df['Extension'] = df['Extension'].apply(lambda x: None if x == '' else x)

df.head(10)
```

```python
print(df.value_counts('Extension'))
print("And None values: ",df['Extension'].isnull().sum())
print("\n",df.value_counts('weird extension'))
print("And None values: ",df['weird extension'].isnull().sum())

print("\nhow many unique files are there?")
print(len(df['Filename'].unique()))
print("how many files are there?")
print(len(df['Filename']))

print("\nAll that have weird extension their real extension is mp3?")
print(df[df['weird extension'].notnull()]['Extension'].value_counts())
```
